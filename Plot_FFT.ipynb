```python
import numpy as np
import datetime 
import matplotlib.pyplot as plt
import seaborn as sns
import os
import pandas as pd
```

```python
datax      = [14, 16, 18, 20, 22]
recur_fft  = [14, 57, 172, 766, 3284]
iter_fft   = [6, 31, 127, 595, 2649]

recur_pfft = [904, 3477, 15156, 73670, 307660] # parallel
leg_list = ["FFT", "PFFT"]

plt.figure()
plt.title("FFT v.s PFFT")
plt.plot(datax, recur_fft, "o-")
plt.plot(datax, iter_fft , 'o-')
plt.xlabel("Power of 2")
plt.ylabel("Time ( ms )")
plt.legend(leg_list)
plt.grid(True)
plt.savefig("pic1.png")
plt.show()
```
